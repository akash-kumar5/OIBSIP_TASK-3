```python
def get_valid_input(prompt, lower_limit=None, upper_limit=None):
    """
    Prompt the user for numeric input and validate within specified limits.
    """
    while True:
        try:
            value = float(input(prompt))
            if (lower_limit is None or value >= lower_limit) and (upper_limit is None or value <= upper_limit):
                return value
            else:
                print("Invalid input. Please enter a valid value.")
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

def calculate_bmi(weight, height):
    """
    Calculate BMI using the formula: BMI = weight / (height ** 2)
    """
    bmi = weight / (height ** 2)
    return bmi

def classify_bmi(bmi):
    """
    Classify BMI into health categories based on predefined ranges.
    """
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

def main():
    print("BMI Calculator")
    print("---------------")

    # Get user input for weight and height with validation
    weight = get_valid_input("Enter your weight in kilograms: ", lower_limit=0.1)
    height = get_valid_input("Enter your height in meters: ", lower_limit=0.1)

    # Calculate BMI
    bmi = calculate_bmi(weight, height)

    # Classify BMI
    category = classify_bmi(bmi)

    # Display results
    print("\nResults:")
    print(f"Your BMI is: {bmi:.2f}")
    print(f"Category: {category}")

if __name__ == "__main__":
    main()
    
```

```python
import requests

def get_weather(api_key, location):
    """
    Get current weather data for a specified location using OpenWeatherMap API.
    """
    base_url = "http://api.openweathermap.org/data/2.5/weather"
    params = {
        "q": location,
        "appid": api_key,
        "units": "metric"  # Use "imperial" for Fahrenheit
    }

    try:
        response = requests.get(base_url, params=params)
        data = response.json()

        if response.status_code == 200:
            return data
        else:
            print(f"Error: {data['message']}")
            return None
    except Exception as e:
        print(f"Error: {e}")
        return None

def display_weather(weather_data):
    """
    Display basic weather information.
    """
    if weather_data:
        print("\nCurrent Weather:")
        print(f"Temperature: {weather_data['main']['temp']}Â°C")
        print(f"Humidity: {weather_data['main']['humidity']}%")
        print(f"Weather Conditions: {weather_data['weather'][0]['description']}")
    else:
        print("Unable to fetch weather data.")

def main():
    print("Basic Weather App")
    print("-----------------")

    # Input your OpenWeatherMap API key here
    api_key = "YOUR_API_KEY"

    # Get user input for location
    location = input("Enter the city or ZIP code: ")

    # Get weather data
    weather_data = get_weather(api_key, location)

    # Display weather information
    display_weather(weather_data)

if __name__ == "__main__":
    main()
```
